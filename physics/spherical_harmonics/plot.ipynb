```python
from scipy.special import sph_harm
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import cmath
```

```python
def orbit(m, l, phy, theta):
    r = projection_real(m, l, phy, theta)

    x = r * np.sin(theta) * np.cos(phy)
    y = r * np.sin(theta) * np.sin(phy)
    z = r * np.cos(theta)
    return x, y, z

def projection_real(m, l, phy, theta):
    Y = sph_harm(m, l, phy, theta)
    Y_pr = sph_harm(m, l, phy, theta)
    Y_mn = sph_harm(-m, l, phy, theta)
    if m == 0:
        return np.abs(Y)
    if m > 0:
        return abs(Y_pr + Y_mn)
    if m < 0:
        return abs(Y_pr - Y_mn)
    


    
def plot(x, y, z, m, l):
    # FigureとAxes
    fig = plt.figure(figsize = (8, 6))
    ax = fig.add_subplot(111, projection='3d')

    # 軸ラベルを設定
    ax.set_xlabel("x", size = 16)
    ax.set_ylabel("y", size = 16)
    ax.set_zlabel("z", size = 16)

    # アングルを設定
    ax.view_init(30, 30)

    # 軌道を描画
    ax.plot_surface(x, y, z, cmap=cm.coolwarm, 
                edgecolor="black", linewidth=0.4)
```

```python
# theta,phiの格子点データ
t = np.linspace(0, np.pi, 257)
p = np.linspace(0, 2*np.pi, 257)

# 格子点の作成
theta, phy = np.meshgrid(t, p)
```

```python
# s軌道
l = 0
for m in range(-l, l+1):

    x, y, z = orbit(m, l, phy, theta)

    # 軌道を描画
    plot(x, y, z, m, l)
```
