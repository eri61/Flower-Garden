```python
import math

from scipy.special import eval_hermite
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import ArtistAnimation
```

```python
def wave_func(n, x):
    norm = np.sqrt(2**n * math.factorial(n) * np.sqrt(np.pi))**(-1)
    H = eval_hermite(n, x)
    wave_func = norm * H * np.exp(-x**2 / 2)
    return wave_func
```

```python
def classic(n):
    return n + 1/2
```

```python
def hermite(n):
    x = np.linspace(-7.5, 7.5, 10000)
    y = wave_func(n, x)

    # plot 
    text = plt.text(-7.5, 0.5, f"n = {n}", va='top', ha='left', fontsize=12)
    frame = plt.plot(x, y, color='royalblue', 
            label='Hermite polynomial')
    return frame + [text]
```

```python
# test 
hermite(12)
```

```python
fig = plt.figure(figsize=(7, 5))
frames = []
ns = range(20)
for i in ns:
    frames.append(hermite(i))

animate = ArtistAnimation(fig, frames, interval=300)
animate.save("./hermite.gif")
```
