```python
import numpy  as np

theta = np.linspace(0, 2 * np.pi, 100)
phi = np.linspace(0, np.pi, 100)
rho = 1 / (1 - 0.5*np.sin(theta)**2)**(3/4)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(projection='3d')

x = rho * np.outer(np.ones_like(phi), np.cos(theta))
y = rho * np.outer(np.cos(phi), np.sin(theta))
z = rho * np.outer(np.sin(phi), np.cos(theta))
ax.plot_wireframe(x, y, z)
```
